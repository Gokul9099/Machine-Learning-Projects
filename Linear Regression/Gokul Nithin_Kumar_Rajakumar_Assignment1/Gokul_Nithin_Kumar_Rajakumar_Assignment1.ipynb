# Linear Regression

        We are finding linear regression for a discrete set of features because to predict an output for a selected set of input features. The linear regression line helps us to predict approx future outputs for a given set of features.
        So in order to test the accuracy we must split the existing dataset into two, 80% for training anf 20% for testing.
        Here we are spliting the dataset RANDOMLY using random permutations and splitting it in the ratio 20:80.
        Later we are storing it into two seperate csv's namely train_set.csv and test_set.csv respectively.

```python
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
from abc import ABC, abstractmethod

# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    
class LinearModel(BaseModel):
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None
        self.w_lms = None

    # check if the matrix is 2-dimensional. if not, raise an exception    
    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):
        """
            train linear model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """        
        pass
    
    @abstractmethod
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """        
        pass 
```

# Least Square
    A mathematical procedure for finding the best-fitting curve to a given set of points by minimizing the sum of the squares of the offsets of the points from the curve. The sum of the squares of the offsets is used instead of the offset absolute values because this allows the residuals to be treated as a continuous differentiable quantity. However, because squares of the offsets are used, outlying points can have a disproportionate effect on the fit, a property which may or may not be desirable depending on the problem at hand.    
    
    How to find best fit line...
    
    Step 1:	For each (x,y) calculate x square and xy
    Step 2:	Sum all x, y, x square and xy, which gives us Σx, Σy, Σx2 and Σxy.
    Step 3:	Calculate Slope m:
        m =  (NΣxy − Σx Σy) / N(Σx2) − (Σx)2 
                where N is the number of points.
    Step 4:	Calculate constant c:
        c =  Σy − m(Σx)N 
    Step 5: Assemble the equation of a line
            y = mx + c

# Least Mean Squares

    Assumption:	The	output	is	a	linear	function	of	the	inputs
            y= b0 +	b1 x1 +	b2 x2
    
    Learning: Using the training data to find the best possible value of w
    
    Prediction: Given the values for x1, x2 for a new period, use the learned w to predict the Temperaure for the new	period of time.
    
    How LMS Works: 
     - Inputs are vectors 
     - Outputs are real numbers:
     - We have a training set	
            D = { (x1,y1),(x2,y2),.....,(xm,ym)}
     - We want to approximate y as	
            y = w1 + w2.x2 +....+ wn.xn
              =	wT.x
              w is the learned weight vector.

# Code Explanation
        Here in the train() method we are training the model in batches so that each time the model gets a new updated weight which is curving towards the actul output value. So for that we are using a for loop to update weights in batches.
        In train_step() method we are calculating weight,
                    w -= alpha * (w.T @ x - t) * x
                  where alpha is the learning rate.
        and in the use() method we are using this calculated weight to calculate the predicted output.

```python
import collections # for checking iterable instance

# LMS class 
class LMS(LinearModel):
    """
        Least Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, alpha):
        LinearModel.__init__(self)
        self.alpha = alpha
    
    # batch training by using train_step function
    def train(self, X, T):
        N= X.shape[0]
        for i in range(N):
            self.train_step(X[i],T[i])
            
    # train LMS model one step 
    # here the x is 1d vector
    def train_step(self, x, t):
        if len(x.shape) != 2:
            x=np.insert(x,0,1).reshape(-1,1)
        if self.w_lms is None:
            self.w_lms = np.zeros((x.shape[0],1))
        y = self.w_lms.T @ x - t
        self.w_lms -= self.alpha * (y) * x
          ## TODO: replace this with your codes
    
    # apply the current model to data X
    def use(self, X):
        if len(X.T.shape) !=2:
            Y = np.insert(x,0,1).reshape(-1,1)
        Y = self.add_ones(X).T
        Y1 = self.w_lms.T @ Y
        return Y1.T  ## TODO: replace this with your codes
```

# Analysis/ Comparison of Algorithms
        Of course Least mean square algorithm is far more better than least square algorithm because in real world the inputs differs periodically. If this happens least square algorithm will calculate weight for the entire dataset. Where as in least mean square algorithm it dynamically calculates weight for every input batch leading to better weight predictions.
        Least square is efficient in terms of calculating the weight whereas least mean square takes more calculation time for caluculating the weight but leads to better results. Thus LMS is better than LS.
